# Import Sentiment Data

```python
# Import data already with datetime class 
df = pd.read_csv("final_with_states_codes.csv", parse_dates = ['date'])

# each day with no tume for better grouping
df['date_day'] = pd.to_datetime(df['date']).dt.date

# dateday as string to better work with it
df['date_day_string'] = df['date_day'].astype(str)

# Which week
df['date_week'] = df['date'].dt.strftime('%Y-%U')

# Which month
df['date_month'] = df['date'].dt.strftime('%Y-%m')
```

```python
dfinal = df[['valid_id', 'date', 'date_day', 'date_week', 'date_month', 'name', \
             'admin1', 'State', 'admin2', 'followers', 'retweets', \
             'replies', 'likes', 'quotes', 'score']]

dfinal = dfinal.rename(columns={"name": "city", "admin1": "state", "admin2": "country", "State": "state_code"})
```
